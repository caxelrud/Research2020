```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%config InlineBackend.figure_format = 'retina'
az.style.use("arviz-darkgrid")
```

#### Code 10.1

```python
d = {
    "A": [0, 0, 10, 0, 0],
    "B": [0, 1, 8, 1, 0],
    "C": [0, 2, 6, 2, 0],
    "D": [1, 2, 4, 2, 1],
    "E": [2, 2, 2, 2, 2],
}
p = pd.DataFrame(data=d)
```

#### Code 10.2

```python
p_norm = p / p.sum(0)
```

#### Code 10.3

```python
def entropy(x):
    y = [0 if i == 0 else i * np.log(i) for i in x]
    return -sum(y)


H = p_norm.apply(entropy, axis=0)
H
```

#### Code 10.4

```python
ways = [1, 90, 1260, 37800, 113400]
logwayspp = np.log(ways) / 10
plt.plot(logwayspp, H, "o")
plt.plot([0.0, max(logwayspp)], [0.0, max(H)], "--k")
plt.ylabel("entropy")
plt.xlabel("log(ways) per pebble");
```

#### Code 10.9

```python
def sim_p(G=1.4):
    x123 = np.random.uniform(size=3)
    x4 = (G * np.sum(x123) - x123[1] - x123[2]) / (2 - G)
    x1234 = np.concatenate((x123, [x4]))
    z = np.sum(x1234)
    p = x1234 / z
    return -np.sum(p * np.log(p)), p
```

#### Code 10.10

```python
H = np.zeros(10 ** 5)
p = np.zeros((10 ** 5, 4))
for rep in range(10 ** 5):
    h, p_ = sim_p()
    H[rep] = h
    p[rep] = p_
```

```python
az.plot_kde(H)
plt.xlabel("Entropy")
plt.ylabel("Density");
```
